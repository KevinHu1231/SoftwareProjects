```python
import numpy as np

v_1 = [0]
v_2 = [0]
w_1 = [0]
w_2 = [0]
x1 = 2
x2 = 1
lr = 1e-4
e = 1e-8
b = 0.9
t = 2

v1 = b*(v_1[0])+(1-b)*(2*x1*(x1*w_1[0]-t))**2
v2 = b*(v_2[0])+(1-b)*(2*x2*(x2*w_2[0]-t))**2
w1 = w_1[0] - lr*(2*x1*(x1*w_1[0]-t))/((v1)**(1/2)+e)
w2 = w_2[0] - lr*(2*x2*(x2*w_2[0]-t))/((v2)**(1/2)+e)

v_1 += [v1]
v_2 += [v2]
w_1 += [w1]
w_2 += [w2]

while (abs(w_1[-1] - w_1[-2]) > 1e-100) and (abs(w_2[-1] - w_2[-2]) > 1e-100):
  v1 = b*(v_1[-1])+(1-b)*(2*x1*(x1*w_1[-1]-t))**2
  v2 = b*(v_2[-1])+(1-b)*(2*x2*(x2*w_2[-1]-t))**2
  w1 = w_1[-1] - lr*(2*x1*(x1*w_1[-1]-t))/((v1)**(1/2)+e)
  w2 = w_2[-1] - lr*(2*x2*(x2*w_2[-1]-t))/((v2)**(1/2)+e)
  v_1 += [v1]
  v_2 += [v2]
  w_1 += [w1]
  w_2 += [w2]

W_adp = np.array([[w_1[-1]],[w_2[-1]]])
print("W_adaptive: ")
print(W_adp)

X = np.array([[2,1]])
W_star = X.T @ (1/(X@X.T)) * t
print("W_solution: ")
print(W_star)
```

```python
import numpy as np

def grad_w(w,x,t):
  return (2*x.T@x@w - 2*t*x).squeeze()

def rmsprop(v,w,x,t,lr,b,e):
  grad = grad_w(w,x,t)
  v = b*v + (1-b)*(grad**2)
  w = w - lr*grad/(v**(1/2)+e)
  return v, w
```

```python
x = np.array([[2,1]])
v = np.array([0.0,0.0])
w = np.array([0.0,0.0])

lr = 1e-4
e = 1e-8
b = 0.9
t = 2

for i in range(10000):
  v, w = rmsprop(v,w,x,t,lr,b,e)

print("RMSProp Solution:", w)

w_star = x.T @ (1/(x@x.T)) * t
print("Actual Solution:", w_star)
```
